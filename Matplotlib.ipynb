```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as ipy
```

```python
def plot_func(frequency):
    t = np.linspace(0, 1, 100)
    plt.figure(2)
    plt.plot(t, np.sin(2 * np.pi * frequency * t))
    plt.ylim(-1.1, 1.1)
    plt.title('Interactive Plot')
    plt.show()

interactive_plot = ipy.interactive(plot_func, frequency=(1, 10, 0.1))
interactive_plot
```
