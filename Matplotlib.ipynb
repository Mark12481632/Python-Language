```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as ipy
```

```python
# Example of interractive plot.

def plot_func(frequency):
    t = np.linspace(0, 1, 100)
    plt.figure(2)
    plt.plot(t, np.sin(2 * np.pi * frequency * t))
    plt.ylim(-1.1, 1.1)
    plt.title('Interactive Plot')
    plt.show()

interactive_plot = ipy.interactive(plot_func, frequency=(1, 10, 0.1))
interactive_plot
```

```python
# 1. Plotting
x_vals = [x for x in np.arange(0, 50, 0.5)]
y_vals_1 = [75 + 10*np.random.normal(size=1) for x in x_vals]
y_vals_2 = [77 + 11*np.random.normal(size=1) for x in x_vals]

plt.plot(x_vals, y_vals_1, color='g', label='Avg. 75', linestyle='--')
plt.plot(x_vals, y_vals_2, color='r', label='Avg. 77', linestyle='-', linewidth=3)
plt.xlabel('X-axis Title')
plt.ylabel('Y-axis Title')
plt.title('Diagram Title')
plt.legend(loc='upper right')
#plt.tight_layout()
```

```python
# Scatter Plot

plt.scatter(x_vals, y_vals_1, color='g', label='Avg. 75', marker='o')
plt.scatter(x_vals, y_vals_2, color='r', label='Avg. 77', marker='x')
plt.xlabel('X-axis Title')
plt.ylabel('Y-axis Title')
plt.title('Diagram Title')
plt.legend(loc='upper right')
plt.grid(True)
```
