```python
import pandas as pd
```

```python
# Set display params for Pandas' output:
# Set the display width to a larger value (e.g., 150 characters)
pd.set_option('display.width', 170)

# Set the maximum number of columns to be displayed (No limit)
pd.set_option('display.max_columns', None)

# Set the maximum number of rows to be displayed (10)
pd.set_option('display.max_rows', 10)

# Set the display options
pd.set_option('display.max_rows', 10)
pd.set_option('display.min_rows', 10)
```

**Loading and displaying data from CSV file.**

```python
# Load the Google Play Store Apps dataset
url = "https://raw.githubusercontent.com/Mark12481632/Python-Language/main/melb_data.csv"

df = pd.read_csv(url)

# Display the first few 6 rows of the dataset
print(df.head(6))

print("===============================================")

# Display last 4 rows of dataset
print(df.tail(4))
```

```python
# Show some statistics on the CSV data:
print("Shape:", df.shape)
print("Columns:", df.columns)

# Summary information on CSV data.
df.info()
```
